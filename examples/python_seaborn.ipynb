### Basic usage

- import colors and templates from oistyle
- pass template to seaborn.set_style()

```python
import seaborn as sns
import matplotlib.pylab as plt
import pandas as pd
import plotly.express as px
from oistyle_python.seaborn import base_template, bar_template, line_template, colors
```

```python
wide_df = px.data.medals_wide()
wide_df = wide_df.set_index('nation').stack().reset_index()
wide_df.columns = ['nation', 'medal', 'amount']


sns.set_style("white", bar_template)

ax = sns.histplot(wide_df, x='nation', hue='medal', weights='amount',
             multiple='stack', palette=colors["darkblue_lightblue_gradient_5"][:3])
ax = ax.set_ylabel('amount')
```
